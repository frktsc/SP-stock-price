```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data_MSFT = pd.read_excel(r"stocksprice-furkanalitasci.xlsx",sheet_name="MSFT")
Data_Tesla = pd.read_excel(r"stocksprice-furkanalitasci.xlsx",sheet_name="Tesla")
Data_Apple = pd.read_excel(r"stocksprice-furkanalitasci.xlsx",sheet_name="Apple")
Data_SP500 = pd.read_excel(r"stocksprice-furkanalitasci.xlsx",sheet_name="SP500")
Data_Combined = pd.read_excel(r"stocksprice-furkanalitasci.xlsx",sheet_name="combined")
```

```python
Data_Tesla.head()
```

```python
Data_Tesla.describe()
```

```python
print(Data_Tesla.info())
```

```python
tesla_skewness = Data_Tesla['Close'].skew()
print("skewness:", tesla_skewness)

teskla_kurtosis = Data_Tesla['Close'].kurtosis()
print("Kurtosis:", teskla_kurtosis)
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=Data_Tesla['Close'], kde=True, bins=50, color='purple', edgecolor='black')

plt.xlabel('Tesla Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Tesla', fontsize=16)
plt.show()
```
