### REGRESSION ANALYSIS FOR TESLA ###

```python
Data_Combined.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
Data_Combined.head()
```

```python
Data_Combined.isna().sum()
```

```python
Data_Combined.dropna(axis=0,inplace=True)
```

```python
X_tesla=Data_Combined['Tesl%chng'].values.reshape(-1,1)
y=Data_Combined['SP%chng']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tesla, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Making predictions on the test set
y_pred = model.predict(X_test)

# Evaluating the model's performance
from sklearn.metrics import mean_squared_error, r2_score
print("Mean squared error (MSE) for Tesla: ", mean_squared_error(y_test, y_pred))
print("Coefficient of determination (R^2) for Tesla: %.2f" % r2_score(y_test, y_pred))

# Comparing the actual values with the predictions
plt.scatter(X_test, y_test, color='black', label='Actual values')
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Predictions')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Model')
plt.legend()
plt.show()
```

```python
covariance = np.cov(Data_Combined['Tesl%chng'], Data_Combined['SP%chng'])
cov_stock_market = covariance[0, 1]
var_market = covariance[1, 1]

beta_tesla = cov_stock_market / var_market
print(f"Tesla Beta Value: {beta_tesla}")
```

```python
# Assuming df is your DataFrame with the stock prices, and it has columns named 'Tesla_Close', 'Microsoft_Close', 'Apple_Close'

# Calculate daily returns
Data_Tesla['Tesla_Return'] = Data_Tesla['Close'].pct_change()
Data_MSFT['Microsoft_Return'] = Data_MSFT['Close'].pct_change()
Data_Apple['Apple_Return'] = Data_Apple['Close'].pct_change()

# Plot histograms
plt.figure(figsize=(18, 5))

# Tesla
plt.subplot(1, 3, 1)  # 1 row, 3 columns, 1st subplot
plt.hist(Data_Tesla['Tesla_Return'].dropna(), bins=30, color='orange')
plt.title('Tesla Daily Returns')
plt.xlabel('Return')
plt.ylabel('Frequency')

# Microsoft
plt.subplot(1, 3, 2)  # 1 row, 3 columns, 2nd subplot
plt.hist(Data_MSFT['Microsoft_Return'].dropna(), bins=30, color='green')
plt.title('Microsoft Daily Returns')
plt.xlabel('Return')

# Apple
plt.subplot(1, 3, 3)  # 1 row, 3 columns, 3rd subplot
plt.hist(Data_Apple['Apple_Return'].dropna(), bins=30, color='red')
plt.title('Apple Daily Returns')
plt.xlabel('Return')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
daily_returns_apple=Data_Apple["Close"].pct_change()
daily_returns_msft=Data_MSFT["Close"].pct_change()
daily_returns_Tesla=Data_Tesla["Close"].pct_change()
```

```python
daily_returns_combined = pd.DataFrame({
    'Apple': daily_returns_apple,
    'Microsoft': daily_returns_msft,
    'Tesla': daily_returns_Tesla
})
```

```python
sns.pairplot(daily_returns_combined.dropna(),kind='reg')
plt.suptitle('Daily Returns: Apple vs. Microsoft vs. Tesla', y=1.02)
plt.show()
```

```python
corr_matrix = daily_returns_combined.corr()

# Heatmap çizimi için farklı bir renk paleti kullanma
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='inferno', vmin=-1, vmax=1)
plt.title('Correlation Matrix with Improved Color Scheme')
plt.show()
```
